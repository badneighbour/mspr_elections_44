```python
import pandas as pd
```

```python

base_column_names = [
    "Code du département", "Libellé du département", "Code de la circonscription",
    "Libellé de la circonscription", "Code de la commune", "Libellé de la commune",
    "Code du b.vote", "Inscrits", "Abstentions", "% Abs/Ins", "Votants", "% Vot/Ins",
    "Blancs", "% Blancs/Ins", "% Blancs/Vot", "Nuls", "% Nuls/Ins", "% Nuls/Vot",
    "Exprimés", "% Exp/Ins", "% Exp/Vot"
]

unnamed_column_names = [
    "N°Panneau", "Sexe", "Nom", "Prénom", "Voix", "% Voix/Ins", "% Voix/Exp"
]

def add_44_to_commune(code_commune):
    return '44' + code_commune.zfill(3)

def process_file(file_path, output_path):
    with open(file_path, 'r', encoding='ISO-8859-1') as file:
        lines = file.readlines()

    processed_lines = []
    for line in lines:
        parts = line.strip().split(';')
        if len(parts) >= 5 and parts[0].strip() == '44':
            parts[4] = add_44_to_commune(parts[4].strip())
            processed_lines.append(parts)

    num_columns = len(processed_lines[0])
    column_names = base_column_names[:]

    num_groups = (num_columns - len(base_column_names)) // len(unnamed_column_names)

    for i in range(num_groups):
        suffix = f"ct{i+1}"
        for base_name in unnamed_column_names:
            column_names.append(f"{base_name}{suffix}")

    remaining_columns = num_columns - len(column_names)
    if remaining_columns > 0:
        for i in range(remaining_columns):
            column_names.append(f"Unnamed_Extra_{i+1}")

    df = pd.DataFrame(processed_lines, columns=column_names)
    df.to_csv(output_path, index=False, encoding='utf-8')

file_path_T2 = 'PR17_BVot_T2_FE.txt'
output_path_T2_csv = 'presidentielle2017_T2_filtre.csv'
process_file(file_path_T2, output_path_T2_csv)

file_path_T1 = 'PR17_BVot_T1_FE.txt'
output_path_T1_csv = 'presidentielle2017_T1_filtre.csv'
process_file(file_path_T1, output_path_T1_csv)
```
